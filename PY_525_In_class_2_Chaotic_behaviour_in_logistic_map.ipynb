# In Class 2: Chaotic behaviour in logistic map


$$x_{n+1} = 1- \lambda x_n^2 $$



Verhulst developed a population model which takes into account that the resources are limited. Without this limitation the population density is described by 
\begin{align}
    N_{n+1} = r N_{n}, 
\end{align}
where $N_n$ is the population density in year $n$ and $r$ is the coefficient characterising the reproduction rate. Verhulst assumed that the reproduction rate depends on the population density as 
\begin{align}
    \frac{r}{r_0} = 1 - \frac{N_n}{C}\,. 
\end{align}
Defining a new variable $y_n = \frac{N_n}{C}$ the model can be rewritten as 
\begin{align}
    y_{n+1} = r_0 y_{n} (1-y_n)\,.  
\end{align}
By shifting and rescaling the variable $x_n = a + b y_n$, we can fix the coefficients $a$ and $b$ such that the model takes its conventional form of the logistic map
\begin{align}
    x_{n+1} = 1 - \lambda x_n^2\,,  
\end{align}
where $\lambda$ is a function of $r_0$.
