# Problem 4: Chaotic behaviour in logistic map

$$x_{n+1} = 1- \lambda x_n^2 $$

```python
C = (0.8666666666666667, 0.5176470588235295, 0.3215686274509804)
```

```python
import matplotlib.pyplot as plt
#import plotmod
import sys 
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import math

def f(x, l):
    return 1-l*x*x 

def orbit(x0,l,N):
    orbit_list = []
    for i in range(N):
        x = f(x0,l)
        orbit_list.append([x0,x])
        orbit_list.append([x,x])
        x0=x
    return (np.array(orbit_list))
    
orbit1 = orbit(0.1,0.5,20)  
```

## Fixed point:

```python
lamb=0.5

orbit1 = orbit(0.1,lamb,20)  

plt.plot(orbit1[:,0][::2],color=C)
plt.plot(orbit1[:,0][::2],'ko')
#plt.xlim(0,1.1)
plt.ylim(0,1.1)
plt.tight_layout()

plt.xlabel(r'$n$')
plt.ylabel(r'$x_{n}$')
```

# Bifurcation Tree

```python
lamb_a = np.arange(0,2,0.01/2)
x_a = []
y_a = []

n_conv=3100

def Lyapunov(x, lamb):
    return (1.0/len(x) * sum (np.log(2 * lamb * abs(x))))

for lamb in lamb_a:
    orbit_s = orbit(0,lamb,n_conv)
    x = np.unique(orbit_s[2*n_conv-200:,0])
    #print(len(x),lamb)
    
    for i in x: 
        x_a.append(i)
    y_a = y_a + [lamb]*len(x)

plt.plot(y_a, x_a,'.',lw=0.1,markersize=0.5,color=(0.8666666666666667, 0.5176470588235295, 0.3215686274509804))
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$x_{f}$')
plt.tight_layout()
plt.savefig('Chaos'+'.pdf')
```

```python
xl_a = []
yl_a = []

for lamb in lamb_a:
    orbit_s = orbit(0,lamb,n_conv)
    x = np.unique(orbit_s[2*n_conv-200:,0])
    #print(len(x),lamb)
    
    l = Lyapunov(x,lamb)
    
    xl_a = xl_a + [l]
    yl_a = yl_a + [lamb]

plt.plot([0,2],[0,0],'--',lw=1)
plt.plot(yl_a, xl_a,'-',lw=1.5,color=C)
plt.ylim(-2,0.5)
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$l$')
plt.tight_layout()

plt.savefig('Lyap'+'.pdf')
```

# Shannon Entropy

```python
xS_a = []
yS_a = []

min_x=-1
max_x=1

NoI = 100 
step = (max_x - min_x) / NoI 

def ShannonEntropy(x):  
  idx  = ((x  - min_x) / step)  
  
  idx = ([int(all) for all in idx])
  set_idx = set([int(all) for all in idx]) 
  p = []
  length = len(set_idx)
  for unq in set_idx:
    p.append(idx.count(unq))
  
  NormP = np.array(p)/sum(p)

  H = - sum( NormP * np.log(NormP)  )
  
  return (H) 

for lamb in lamb_a:
    orbit_s = orbit(0,lamb,n_conv)
    x = np.unique(orbit_s[2*n_conv-400:,0][::2])
    #print(len(x),lamb)
    
    H = ShannonEntropy(x)
    
    xS_a = xS_a + [H]
    yS_a = yS_a + [lamb]
```

```python
plt.plot(yS_a, xS_a,color=C)
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$H$')
```

# RG equation, evolution operator 
$f_{k+1} (x)  = \alpha_k f_k \left( f_k \left(\frac{x}{\alpha_k} \right)\right)\,,\quad 
    \alpha_k = \frac{1}{f_k(f_k(0))}$ with $f_0(x) = 1 - \lambda x^2$

```python
def f(k,x,lamb):
    if(k>0):
        alpha = 1/f(k-1,f(k-1,0,lamb),lamb)
        result=alpha*f(k-1,f(k-1,x/alpha,lamb),lamb)
    else: 
        result = 1 - lamb*x*x
    return result; 

def data(k,lamb):
    xa = np.arange(-1,1,0.01)
    return (xa, np.array([f(k,x,lamb) for x in xa]))
```

```python
for k in [0,1,2,3,4,5]:
    (x,y) = data(k,1.40115518909205060052382-0.001) 
    plt.plot(x,y)
```

```python
for k in [0,1,2,3,4,5]:
    (x,y) = data(k,1.40115518909205060052382+0.001) 
    plt.plot(x,y)
```
